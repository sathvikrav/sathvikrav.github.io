```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv('android-games.csv')
```

```python
df = df.rename(columns={'growth (30 days)': 'Percent growth in 30 days', 'growth (60 days)': 'Percent growth in 60 days'})
```

```python
for index, row in df.iterrows():
    array = row['installs'].split(' ')
    
    if array[1] == 'M':
        number = float(array[0])
        number *= 1000000.0
    else:
        number = float(array[0])
        number *= 1000.0
        
    df.at[index, 'installs'] = number
    
df['installs'] = df['installs'].astype('float')
```

```python
df = df.drop_duplicates(subset=['title'])
```

```python
for index, row in df.iterrows():
    df.at[index, '% 5 star'] = row['5 star ratings'] / row['total ratings']
    df.at[index, '% 4 star'] = row['4 star ratings'] / row['total ratings']
    df.at[index, '% 3 star'] = row['3 star ratings'] / row['total ratings']
    df.at[index, '% 2 star'] = row['2 star ratings'] / row['total ratings']
    df.at[index, '% 1 star'] = row['1 star ratings'] / row['total ratings']
```

```python
res = stats.linregress(df['rank'].values, df['Percent growth in 60 days'].values)
print(res)
```

```python
lm = linear_model.LinearRegression()

# reshape the array so that the model understands that our independent variable is entered in 2 dimensions
ind = df['rank'].values.reshape(-1, 1)
dep = df['Percent growth in 60 days'].values

lm.fit(ind, dep)

print(lm.coef_)
print(lm.intercept_)
```

```python
plt.subplots(figsize=(10,10))

y = 'Percent growth in 60 days'
x = 'total ratings'
plt.scatter(x=df[x].values, y=df[y].values)
plt.xlabel(x)
plt.ylabel(y)
plt.title(y + ' vs. ' + x)
```

```python
plt.subplots(figsize=(10,10))


temp_df = pd.DataFrame()
temp_df['col1'] = df['total ratings'].astype(float)
temp_df['col2'] = np.reciprocal(df['Percent growth in 60 days'])

indices = []

for index, row in temp_df.iterrows():
    if np.isnan(row['col1']) or np.isnan(row['col2']) or np.isinf(row['col1']) or np.isinf(row['col2']):
        indices.append(index)
        
temp_df = temp_df.drop(index=indices)

# temp_df

y = 'Reciprocal of Percent growth in 60 days'
x = 'total ratings'
plt.scatter(x=temp_df['col1'].values, y=temp_df['col2'].values)
plt.xlabel(x)
plt.ylabel(y)
plt.ylim([0, 11])
plt.title(y + ' vs. ' + x)

ind = temp_df['col1'].values.reshape(-1,1)
lm.fit(ind, temp_df['col2'])
plt.plot(ind, lm.predict(ind), color='red')

print('Linear regression model given by: y = ' + str(lm.coef_[0]) + 'x + ' + str(lm.intercept_) )
```

```python
plt.subplots(figsize=(30, 20))

grouped = df.groupby('category')
categories = []
residuals = []

for group in grouped:
    categories.append(group[0])
    residuals.append(group[1]['Percent growth in 60 days'].values - lm.predict(group[1]['rank'].values.reshape(-1, 1)))
    
plt.boxplot(x=residuals, labels=categories)
plt.ylabel('Residuals')
plt.title('Residuals vs. Genre')
plt.ylim([-50, 50])
```
