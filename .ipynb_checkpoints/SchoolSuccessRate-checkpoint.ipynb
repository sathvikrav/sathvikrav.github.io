```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
from matplotlib.cm import ScalarMappable
import requests
import bs4
import re
import statsmodels.api as sm
```

```python
df = pd.read_csv('data_1 - data_1.csv')
df2 = pd.read_csv('data_2.tsv', sep='\t')
df3 = pd.read_csv('data_3.tsv', sep='\t')
df4 = pd.read_csv('data_4.tsv', sep='\t')
df5 = pd.read_csv('data_5.tsv', sep='\t')
df6 = pd.read_csv('data_6.tsv', sep='\t')
df7 = pd.read_csv('data_7.tsv', sep='\t')
df8 = pd.read_csv('data_8.tsv', sep='\t')
df9 = pd.read_csv('data_9.tsv', sep='\t')
df2.columns = df.columns
df3.columns = df.columns
df4.columns = df.columns
df5.columns = df.columns
df6.columns = df.columns
df7.columns = df.columns
df8.columns = df.columns
df9.columns = df.columns
df = df.append(df2)
df = df.append(df3)
df = df.append(df4)
df = df.append(df5)
df = df.append(df6)
df = df.append(df7)
df = df.append(df8)
df = df.append(df9)
```

```python
df = df.drop(columns=['category_1', 'category_2', 'category_3'])
```

```python
df = df.set_index('event_id')
```

```python
indices = []

for index, row in df.iterrows():
    if row['event_name'] != 'search_locations' and row['event_name'] != 'view_screen':
        indices.append(index) 
        
df = df.drop(index=indices)
```

```python
indices = []

for index, row in df.iterrows():
    if row['event_name'] == 'view_screen' and row['event_value'] != 'location_map':
        indices.append(index) 
        
df = df.drop(index=indices)
```

```python
df['session_id'] = df['session_id'].astype('int64')
```

```python
indices = []

for index, row in df.iterrows():
    if pd.isna(row['event_time']) or pd.isna(row['school_uuid']):
        indices.append(index)
        
df = df.drop(index=indices)
```

```python
df['event_time'] = df['event_time'].astype('int64')
```

```python
for index, row in df.iterrows():
    df.at[index, 'datetime'] = dt.datetime.utcfromtimestamp(row['event_time']).date()
```

```python
grouped = df.groupby('user_uuid')

queries_per_user = []
success_rate_per_user = []
time_delta_per_user = []
user_uuids = []
sessions_per_user = []

for group in grouped:
    group[1].sort_values(by='datetime', inplace=True)
    
    num_queries = 0
    success_queries = 0
    idx = 0
    
    for index, row in group[1].iterrows():
        if row['event_name'] == 'search_locations' and idx + 1 < len(group[1]) and group[1].iloc[idx + 1]['session_id'] == 
            row['session_id'] and group[1].iloc[idx + 1]['event_name'] == 'view_screen':
            success_queries += 1
            num_queries += 1
        elif row['event_name'] == 'search_locations':
            if idx > 0 and group[1].iloc[idx - 1]['event_name'] == 'search_locations' and 
                group[1].iloc[idx - 1]['session_id'] == row['session_id']:
                if idx == len(group[1]) - 1:
                    num_queries += 1
                elif group[1].iloc[idx + 1]['session_id'] != row['session_id']:
                    num_queries += 1     
        idx += 1
    
    if num_queries > 0:
        user_uuids.append(group[0])
        queries_per_user.append(num_queries)
        success_rate_per_user.append((success_queries / num_queries))
        sessions_per_user.append(group[1]['session_id'].nunique())
        time_delta_per_user.append((group[1].iloc[len(group[1]) - 1]['datetime'] - group[1].iloc[0]['datetime']).days)
    
```

```python
success_rate_per_user
```

```python
user_success_rate = dict()

for i in range(len(user_uuids)):
    user_success_rate[user_uuids[i]] = success_rate_per_user[i]
```

```python
grouped = df.groupby('school_uuid')
avg_success_rate_per_school = list()
school_uuids = list()
unique_user_uuids = set()

for group in grouped:
    sum_schools = 0
    num_users = 0
    
    for index, row in group[1].iterrows():
        if row['user_uuid'] not in unique_user_uuids and row['user_uuid'] in user_success_rate:
            sum_schools += user_success_rate[row['user_uuid']]
            unique_user_uuids.add(row['user_uuid'])
            num_users += 1
        
    if num_users > 0:
        school_uuids.append(group[0])
        avg_success_rate_per_school.append(sum_schools / num_users)
```

```python
school_success_rate = dict()

for i in range(len(avg_success_rate_per_school)):
    if avg_success_rate_per_school[i] not in school_success_rate:
        school_success_rate[avg_success_rate_per_school[i]] = [int(school_uuids[i])]
    else:
        school_success_rate[avg_success_rate_per_school[i]].append(int(school_uuids[i]))
```

```python
users_to_num_searches = dict()

for i in range(len(user_uuids)):
    users_to_num_searches[user_uuids[i]] = queries_per_user[i]
```

```python
grouped = df.groupby('school_uuid')

searches_by_school = dict()
unique_users = set()

all_users = set()

for user in user_uuids:
    all_users.add(user)

for group in grouped:
    num_queries = 0
    
    for index, row in group[1].iterrows():
        if row['user_uuid'] not in unique_users and row['user_uuid'] in all_users:
            num_queries += users_to_num_searches[row['user_uuid']]
            unique_users.add(row['user_uuid'])
        
    searches_by_school[int(group[0])] = num_queries

searches_by_school
```

```python
colleges = pd.read_csv('schools - schools.csv')
```

```python
college_info = {}

idx = 1

for index, row in colleges.iterrows():
    array = row['primary_name'].lower().split(' ')
    url = 'https://www.google.com/search?q='
    
    for ele in array:
        url += ele + '+'
    
    url += 'acres'
    
    res = requests.get(url)
    
    soup = bs4.BeautifulSoup( res.text, "html.parser" )
    
    m = re.search(r'(\d+\.*\d*)\s*acre', soup.text.replace(',', '').replace('-', ''))
    
    number = re.search(r'\d+\.*\d*', m.group(0))
    
    college_info[index + 1] = [row['primary_name'], float(number.group(0))]
    
```

```python
college_students = {}
idx = 1

for index, row in colleges.iterrows():
    array = row['primary_name'].lower().split(' ')
    url = 'https://www.google.com/search?q='
    
    for ele in array:
        url += ele + '+'
    
    url += 'number+of+students'
    
    res = requests.get(url)
    
    soup = bs4.BeautifulSoup( res.text, "html.parser" )
    
    m = re.search(r'\d+,\d+', soup.text)
    
    college_students[idx] = [int(m.group(0).replace(',', ''))]

    idx += 1
```

```python
avg_time_delta_school = dict()
grouped = df.groupby('school_uuid')
unique_users = set()

for group in grouped:
    
    users = group[1].groupby('user_uuid')
    total_time_delta = 0
    num_users = 0
    
    for user in users:
        
        if user[0] not in unique_users:
            unique_users.add(user[0])
            user[1].sort_values(by='datetime')
            total_time_delta += (user[1].iloc[len(user[1]) - 1]['datetime'] - user[1].iloc[0]['datetime']).days
            num_users += 1
            
    avg_time_delta_school[int(group[0])] = total_time_delta / num_users
        
        
avg_time_delta_school
```

```python
# Number of students vs. avg time delta
num_students = list()
time_delta = list()

for key in college_students:
    num_students.append(college_students[key])
    time_delta.append(avg_time_delta_school[key])
    
    
```
