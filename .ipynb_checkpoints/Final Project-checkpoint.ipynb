# Understanding and Predicting the Growth of Mobile Games

## Introduction

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# This dataset contains the top games on the Google Play Store.

df = pd.read_csv('android-games.csv')
```

```python
df
```

```python
df['growth (30 days)'].sum() / 1730
```

Since the paid column and price column are repetitive, we will drop the paid column since it does not give us any added information.

```python
df = df.drop(columns=['paid'])
```

Since the installs column is recognized as an object type, we need to convert our values to numbers

```python
for index, row in df.iterrows():
    array = row['installs'].split(' ')
    number = float(array[0])
    number *= 1000000.0   
    df.at[index, 'installs'] = number
```

```python
df['installs'] = df['installs'].astype('float')
```

```python
df = df.drop_duplicates(subset=['title'])
df
```

```python
df
```

```python
plt.subplots(figsize = (12,8))
plt.scatter(x=df['installs'].values, y=df['Percent growth in 60 days'].values)
x = "Installations"
y = "Percent Growth in 60 days"
plt.xlabel(x)
plt.ylabel(y)
plt.ylim([-10,100])
plt.title(y + " vs. " + x)
plt.show()
```

```python
for index, row in df.iterrows():
    array = row['installs'].split(' ')
    
    if array[1] == 'M':
        number = float(array[0])
        number *= 1000000.0
    else:
        number = float(array[0])
        number *= 1000.0
        
    df.at[index, 'installs'] = number
```

Let's add some more clarity to our dataframe by clarifying what the growth value means.

```python
df = df.rename(columns={'growth (30 days)': 'Percent growth in 30 days', 'growth (60 days)': 'Percent growth in 60 days'})
```

```python
# Convert categorical variable of game category to numerical variable

df2 = pd.get_dummies(df['category']).drop(columns=['GAME WORD'])

for col in df2.columns:
    df[col] = df2[col]
    
df
```

```python
plt.figure(figsize=(30,25))
sns.heatmap(df.corr(), annot=True, cmap='Blues')
plt.show()
```

```python
df.columns
```

```python
y_var = df[["Percent growth in 60 days"]]
x_var = sm.add_constant(df[['rank', 'GAME EDUCATIONAL', 'GAME MUSIC', 'average rating', 'GAME ACTION',
       'GAME ADVENTURE', 'GAME ARCADE', 'GAME BOARD', 'GAME CARD',
       'GAME CASINO', 'GAME CASUAL', 'GAME EDUCATIONAL', 'GAME MUSIC',
       'GAME PUZZLE', 'GAME RACING', 'GAME ROLE PLAYING', 'GAME SIMULATION',
       'GAME SPORTS', 'GAME STRATEGY', 'GAME TRIVIA']])
LinearModel = sm.OLS(y_var, x_var, missing='drop').fit()
print(LinearModel.summary())
```

```python
lm = linear_model.LinearRegression()

ind = df[['rank']].values
dep = df['Percent growth in 60 days'].values

lm.fit(ind, dep)

print(lm.coef_[0])
print(lm.intercept_)
```

```python
# Plot the residual plot
fig, ax = plt.subplots(figsize=(10,10))

predicted_values = (df['rank'].values * lm.coef_[0]).transpose() + (np.ones(shape=(len(df['rank']),)) * lm.intercept_).transpose()
residuals = df['Percent growth in 60 days'].values - predicted_values

ax.scatter(x=df['rank'].values, y=residuals)
ax.set_xlabel('Rank of the App on Play Store')
ax.set_ylabel('Residuals')
ax.set_title('Residuals vs. Rank')
ax.set_ylim([-50,50])
```

```python
scaled_features = StandardScaler().fit_transform(df[['rank', 'total ratings', 'installs', 'average rating',
       'Percent growth in 30 days', 'Percent growth in 60 days', 'price',
       '5 star ratings', '4 star ratings', '3 star ratings',
       '2 star ratings', '1 star ratings', 'GAME ACTION', 'GAME ADVENTURE',
       'GAME ARCADE', 'GAME BOARD', 'GAME CARD', 'GAME CASINO', 'GAME CASUAL',
       'GAME EDUCATIONAL', 'GAME MUSIC', 'GAME PUZZLE', 'GAME RACING',
       'GAME ROLE PLAYING', 'GAME SIMULATION', 'GAME SPORTS', 'GAME STRATEGY',
       'GAME TRIVIA']].values)
scaled_features_df = pd.DataFrame(scaled_features, index=df.index, columns=['rank', 'total ratings', 'installs', 'average rating',
       'Percent growth in 30 days', 'Percent growth in 60 days', 'price',
       '5 star ratings', '4 star ratings', '3 star ratings',
       '2 star ratings', '1 star ratings', 'GAME ACTION', 'GAME ADVENTURE',
       'GAME ARCADE', 'GAME BOARD', 'GAME CARD', 'GAME CASINO', 'GAME CASUAL',
       'GAME EDUCATIONAL', 'GAME MUSIC', 'GAME PUZZLE', 'GAME RACING',
       'GAME ROLE PLAYING', 'GAME SIMULATION', 'GAME SPORTS', 'GAME STRATEGY',
       'GAME TRIVIA'])
scaled_features_df
```

```python
y_var = scaled_features_df[["Percent growth in 60 days"]]
x_var = sm.add_constant(scaled_features_df[['rank']])
LinearModel = sm.OLS(y_var, x_var, missing='drop').fit()
print(LinearModel.summary())
```
