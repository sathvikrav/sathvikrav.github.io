```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from scipy import stats
import statsmodels.api as sm
```

```python
data = pd.read_csv("gap.tsv", sep='\t')
data.head()
```

```python
# Part 1 Exercise 1

plt.scatter(data['year'], data['lifeExp'])
plt.xlabel('year')
plt.ylabel('life expectancy (in years)')
plt.title('Life Expectancy vs. Year')
```

```python
# Part 1 Question 1

# There is a steady increasing linear trend across the years.
```

```python
# Part 1 Question 2

grouped = data.groupby('year')
years = list()
life_exp_per_year = list()

for group in grouped:
    # for each year add all of the life expectancies for the different countries
    
    years.append(group[1]['year'].iloc[0])
    life_exp_per_year.append(group[1]['lifeExp'].values)
    
plt.violinplot(life_exp_per_year,years,widths=4,showmeans=True)
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.title("Life Expectancy Violin Plot")
```

```python
# Part 1 Question 2

# For 1952-1957, the distribution of life expectancy was skewed to the left and unimodal. In 1962-1967, the distribution was
# skewed to the left and bimodal. In 1972, the distribution was skewed to the right and bimodal. From 1977-2007, the 
# distribution was skewed to the right and unimodal. 
```

```python
# Part 1 Question 3

# Depending on the R-squared value, which determines correlation, and the slope of the line, which affects the P value 
# depending upon how much it deviates from the null hypothesis, which says that the slope is 0, I will decide whether to 
# reject the null hypothesis. Since the violin plot shows a drastic change in the skewness of the age distribution over
# time, I would likely reject the null hypothesis that there is no linear relationship between the Year and the Life
# Expectancy. 
```

```python
# Part 1 Question 4

# Since the residuals should be independent from any of the dimensions, they should follow a random pattern when plotted 
# and thus, should have a normal distribution centered at 0 for each of the years in which data was collected.
```

```python
# Part 1 Question 5

# The linear model assumes that the residuals should be independent of both the year and life expectancy dimensions and
# that the residuals should be normally distributed. Therefore, each year should follow a normal distribution of points
# and the variance should be constant, meaning that there should be approximately an equal number of points above and below
# 0, or the median point of the distribution. 
```

```python
# Part 1 Exercise 2

lm = linear_model.LinearRegression()

# reshape the array so that the model understands that our independent variable is entered in 2 dimensions
ind = data['year'].values.reshape(-1, 1)
dep = data['lifeExp'].values

lm.fit(ind, dep)

print(lm.coef_)
print(lm.intercept_)
```

```python
# Part 1 Question 6

# On average, life expectancy increases by about 0.32590383 years per year.
```

```python
# Part 1 Question 7

# Let's find the pvalue using the scipy linear regression function

res = stats.linregress(data['year'].values, data['lifeExp'].values)
print(res.pvalue)
```

```python
# Since the p-value is less than 0.05, which is our designated alpha value for significance testing, this tells us that our 
# data is highly unlikely to have occurred under the null hypothesis and thus, we reject our null hypothesis.
```

```python
# Part 1 Exercise 3 Residual Plot vs. Year

# Use the linear model from earlier to get predictions on our data and add it to our dataframe

data['Predicted lifeExp'] = lm.predict(data['year'].values.reshape(-1, 1))

grouped = data.groupby('year')
years = []
residuals = []

for group in grouped:
    years.append(group[0])
    residuals.append((group[1]['lifeExp'].values) - (group[1]['Predicted lifeExp'].values))


plt.violinplot(residuals,years,widths=4,showmeans=True)
plt.xlabel("Year")
plt.ylabel("Residuals")
plt.title("Residuals vs. Year Plot")
```

```python
# Part 1 Question 8

# No, the results of Exercise 3 do not match my expectations from Question 4. 
```

```python
# Part 1 Exercise 4

grouped = data.groupby('continent')
continents = []
residuals = []

for group in grouped:
    continents.append(group[0])
    residuals.append(group[1]['lifeExp'].values - group[1]['Predicted lifeExp'].values)
    
plt.boxplot(x=residuals, labels=continents)
plt.ylabel('Residuals')
plt.title('Residuals vs. Continent')
```

```python
# Part 1 Question 9

# There is a dependence between model residual and continent. When we are performing a regression analysis of life
# expectancy across time, this would suggest that our model will be untrustworthy unless we account for this dependence since
# each step in this dimension (Continents) has a tangible affect on the residuals. This means our residuals will not be 
# normally distributed and our model will be less accurate as a result. 
```

```python
# Part 1 Exercise 5

# We have 5 continents to create plots for

fig, axs = plt.subplots(5)

grouped = data.groupby('continent')

idx = 0

for group in grouped:
    ind = group[1]['year'].values.reshape(-1, 1)
    dep = group[1]['lifeExp'].values
    
    lm.fit(ind, dep)
    
    # Plot life expectancy for each continent
    
    axs[idx].scatter(ind, dep)
    axs[idx].set_xlabel('Year')
    axs[idx].set_ylabel('Life Expectancy')
    axs[idx].set_title("Life Expectancy vs. Year in " + group[0])
    
    axs[idx].plot(ind, lm.predict(ind), color='red')
    
    axs[idx].figure.set_size_inches(30,30)
    
    idx += 1
```

```python
# Part 1 Question 10

# The regression model should include an interaction term for continent and year because even though all the regression
# models have increasing slope and we can have OK predictive power by not accounting for continent, we can improve our
# predictive accuracy by adding another dimension since the slopes of the five lines vary significantly between the five
# continents. 
```

```python
# Part 1 Exercise 6

data = data.join(pd.get_dummies(data['continent'], prefix='continent'))
data
```

```python
# We will use as few columns as possible so as to not compromise the accuracy of our model. 

ind = data[['year', 'continent_Africa', 'continent_Americas', 'continent_Asia', 'continent_Europe']].values
dep = data['lifeExp'].values

lm2 = linear_model.LinearRegression()
lm2.fit(ind, dep)

print(lm2.coef_)
print(lm2.intercept_)
```

```python
# Part 1 Question 11

# We will calculate the P-value to see the probability that the parameters we found were measured solely by chance.

YVar = data[['lifeExp']]
XVar = sm.add_constant(data[['year', 'continent_Africa', 'continent_Americas', 'continent_Asia', 'continent_Europe']])
LinearModel = sm.OLS(YVar, XVar, missing='drop').fit()
print(LinearModel.summary())
```

```python
# Part 1 Question 11

# From the linear model summary, we can see that the P-value for the coefficients for the y-intercept, year, continent_Africa,
# continent_Americas, and continent_Asia were all below 0.05 (not significantly different from 0), which means that there
# is a low chance that these coefficients were measured solely by chance. 
```

```python
# Part 1 Question 12

print(lm2.coef_)
print(lm2.intercept_)

# For Africa, the equation is f = 0.325a - 25.46087b - 570.8004184
# For Americas, the equation is f = 0.325a - 9.66747167c - 570.8004184
# For Asia, the equation is f = 0.325a - 14.26d - 570.8004184
# For Europe, the equation is f = 0.325a - 2.42e - 570.8004184
# For Oceania, the equation is f = 0.325a - 570.8004184 and the average increase is 0.326 years per year.
```

```python
# Part 1 Exercise 7

data['predicted life_exp'] = lm2.predict(data[['year', 'continent_Africa', 'continent_Americas', 'continent_Asia', 'continent_Europe']].values)

grouped = data.groupby('year')
years = []
residuals = []

for group in grouped:
    years.append(group[0])
    residuals.append((group[1]['lifeExp'].values) - (group[1]['predicted life_exp'].values))

plt.violinplot(residuals,years,widths=4,showmeans=True)
plt.xlabel("Year")
plt.ylabel("Residuals")
plt.title("Residuals vs. Year Plot")
```

```python
# The residual plot perfectly matches the assumptions for a linear model since the residuals are normally distributed, 
# have a constant variance (which is shown by the fact that there are approximately the same number of points above 
# and below the median), and the residuals are clearly independent from the continent and year dimensions. 
```
